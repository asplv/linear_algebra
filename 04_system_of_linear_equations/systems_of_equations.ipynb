$$\tilde{A}=\begin{pmatrix}

Найти соотношение между параметрами $a$, $b$ и $c$, при которых система является несовместной.

Вычтем удвоенную первую строку из второй и утроенную первую строку из третьей строки:

__1.__ Решить систему уравнений методом Крамера:

a) $\begin{cases}
x_{1}-2x_{2}=1 \\
3x_{1}-4x_{2}=7
\end{cases}$

Найдем определитель:

$\begin{vmatrix}
1 & -2 \\ 
3 & -4 \\ 
\end{vmatrix} = 1 \cdot (-4) - 3 \cdot (-2) = -4 + 6 = 2$, $\Delta \ne 0$ => система уравнений определена, т.е. совместна.

Найдем $detA_{1}$, $detA_{2}$:

$x_{i}=\frac{detA_{i}}{detA}$

$detA_{1} = \begin{vmatrix}
1 & -2 \\ 
7 & -4 \\ 
\end{vmatrix} = -4 + 14 = -10$ => $x_1 = \frac{10}{2} = 5$

$detA_{2} = \begin{vmatrix}
1 & 1 \\ 
3 & 7 \\ 
\end{vmatrix} = 7 - 3 = 4$ => $x_2 = \frac{4}{2} = 2$

б) $\begin{cases}
2x_{1}-x_{2}+5x_{3}=10 \\
x_{1}+x_{2}-3x_{3}=-2 \\
2x_{1}+4x_{2}+x_{3}=1
\end{cases}$

Найдем определитель:

$\begin{vmatrix}
2 & -1 & 5 \\ 
1 & 1 & -3 \\ 
2 & 4 & 1
\end{vmatrix} = 2(1+12) + (1+6) + 5(4-2) = 26 + 7 + 10 = 43$, $\Delta \ne 0$ => система уравнений определена, т.е. совместна.

$detA_{1} = \begin{vmatrix}
10 & -1 & 5\\ 
-2 & 1 & -3\\ 
1 & 4 & 1 
\end{vmatrix} = 10(1+12) + 1(-2+3) +5(-8-1) = 130 + 1 - 45 = 86$
 => $x_1 = \frac{86}{43}$

$detA_{2} = \begin{vmatrix}
2 & 10 & 5\\ 
1 & -2 & -3\\ 
2 & 1 & 1 
\end{vmatrix} = 2(-2+3) - 10(1+6) + 5(1+4) = 2 - 70 + 25 = -43$ => $x_2 = -\frac{43}{43} = -1$

$detA_{3} = \begin{vmatrix}
2 & -1 & 10\\ 
1 & 1 & -2\\ 
2 & 4 & 1 
\end{vmatrix} = 2(1+8) + 1(1+4) + 10(4-2) = 18 + 5 + 20 = 43$ => $x_3 = \frac{43}{43} = 1$

__3*.__ Решить систему линейных уравнений методом $LU$-разложения

$$\begin{cases}
2x_{1}+x_{2}+3x_{3}=1 \\
11x_{1}+7x_{2}+5x_{3}=-6 \\
9x_{1}+8x_{2}+4x_{3}=-5
\end{cases}$$

Матрица коэффицентов:

$\begin{pmatrix}
2 & 1 & 3 \\ 
11 & 7 & 5 \\ 
9 & 8 & 4 \\
\end{pmatrix}$

Отнимем от второй строки первую строку, умноженную на 5.5, и от третьей строки первую, умноженную на 4.5:

$\begin{pmatrix}
2 & 1 & 3 \\ 
0 & 3/2 & -11.5 \\ 
0 & 7/2 & -9.5 \\
\end{pmatrix}$

Отнимем от третьей строки вторую, умноженную на $\frac{7}{3}$:

$\begin{pmatrix}
2 & 1 & 3 \\ 
0 & 1.5 & -11.5 \\ 
0 & 0 & 52/3 \\
\end{pmatrix}$


L = $\begin{pmatrix}
1 & 0 & 0 \\ 
5.5 & 1 & 0 \\ 
4.5 & 7/3 & 1 \\
\end{pmatrix}$

U = $\begin{pmatrix}
2 & 1 & 3 \\ 
0 & 1.5 & -11.5 \\ 
0 & 0 & 52/3   \\
\end{pmatrix}$

```python
A = np.array([
    [2, 1, 3],
    [11, 7, 5],
    [9, 8, 4]
])

L = np.array([
    [1, 0, 0],
    [5.5, 1, 0],
    [4.5, 7/3, 1]
])


U = np.array([
    [2, 1, 3],
    [0, 1.5, -11.5],
    [0, 0, 52/3]
])

print(f'Матрица A: \n {A}')
print(f'Произведение матрицы L и U: \n {np.dot(L, U)}')
```

$Ly = b$

$\begin{pmatrix}
1 & 0 & 0 \\ 
5.5 & 1 & 0 \\ 
4.5 & 7/3 & 1 \\
\end{pmatrix} \cdot \begin{pmatrix}
y_1 \\ 
y_2 \\ 
y_3
\end{pmatrix} = \begin{pmatrix}
1 \\ 
-6 \\ 
- 5
\end{pmatrix}$

$\begin{cases}
y_1=1, \\

4.5y_{1} + \frac{7}{3}y_{2} + y_3 = - 5;
\end{cases}$

$\begin{cases}
y_1=1, \\
y_{2} =-11.5, \\
4.5 - \frac{7}{3} \cdot 11.5 + y_3 = - 5;
\end{cases}$

$\begin{cases}
y_1=1, \\
y_{2} =-11.5, \\
y_3 = \frac{52}{3};
\end{cases}$

$Ux = y$

$\begin{pmatrix}
2 & 1 & 3 \\ 
0 & 1.5 & -11.5 \\ 
0 & 0 & 52/3   \\
\end{pmatrix} \cdot \begin{pmatrix}
x_1 \\
x_2 \\
x_3
\end{pmatrix} = \begin{pmatrix}
1 \\ 
-11.5 \\ 
52/3
\end{pmatrix}$

$\begin{cases}
2x_1 + x_2 + 3x_3 =1, \\

\frac{52}{3}x_3 = \frac{52}{3};
\end{cases}$

$\begin{cases}
x_3 = - 1, \\
x_2 = 0, \\
x_3 = 1;
\end{cases}$
