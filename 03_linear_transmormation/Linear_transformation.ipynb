## Тема: Линейные преобразования

__1.__ Найти собственные векторы и собственные значения для линейного оператора, заданного матрицей

$$A=\begin{pmatrix}
-1 & -6\\ 
2 & 6
\end{pmatrix}.$$

Получим характеристическое уравнение:

$\begin{pmatrix}
-1 & -6\\ 
2 & 6
\end{pmatrix} \cdot 
\begin{pmatrix}
x\\
y
\end{pmatrix} = 
\lambda \cdot \begin{pmatrix}
x\\
y
\end{pmatrix}$

$\begin{pmatrix}
-x - 6y \\ 
2x + 6y 
\end{pmatrix} = 
\begin{pmatrix}
\lambda x\\
\lambda y
\end{pmatrix}$

$\begin{cases}
   -x - 6y = \lambda x; \\
   2x + 6y = \lambda y; \\
 \end{cases}$
 
 $\begin{cases}
   -x - 6y - \lambda x = 0; \\
   2x + 6y - \lambda y = 0; \\
 \end{cases}$
 
 $\begin{cases}
   x(-1 - \lambda) - 6y = 0; \\
   2x + y(6 - \lambda) = 0; \\
 \end{cases}$
 
$\begin{vmatrix}
-1 - \lambda & -6\\ 
2 & 6 - \lambda
\end{vmatrix} = 0 $

$(-1 - \lambda)(6 - \lambda) +12 = 0$

$ -6 + \lambda -6\lambda + \lambda^2 +12 = 0$

$ \lambda^2 -5\lambda + 6 = 0$

$\lambda_1 = \frac{5+\sqrt{25 -24}}{2} = 3$

$\lambda_2 = \frac{5-\sqrt{25-24}}{2} = 2$

Собственные значения: $\lambda_1 = 3, \lambda_2 = 2$

Найдем собственные вектора:

1) $\lambda_1 = 3$

$\begin{cases}
  x(-1 - 3) - 6y = 0; \\
  2x + y(6 - 3) = 0; \\
\end{cases}$
 
$\begin{cases}
  -4x - 6y = 0; \\
  2x + 3y = 0; \\
\end{cases}$ => $x = -\frac{3}{2}y$

Возьмем вектор: 
$\begin{pmatrix}
3\\
-2
\end{pmatrix}$

Проверим, подставив значения вектора в уравнение:

$\begin{cases}
  -4\cdot3 + -6\cdot(-2) = 0; \\
  2\cdot3 + 3\cdot(-2) = 0; \\
\end{cases}$

2) $\lambda_2 = 2$

$\begin{cases}
  x(-1 - 2) - 6y = 0; \\
  2x + y(6 - 2) = 0; \\
\end{cases}$
 
$\begin{cases}
  -3x - 6y = 0; \\
  2x + 4y = 0; \\
\end{cases}$ => $x = -2y$


Возьмем вектор: 
$\begin{pmatrix}
2\\
-1
\end{pmatrix}$

Проверим, подставив значения вектора в уравнение:

$\begin{cases}
  (-3)\cdot2 - 6\cdot(-1) = 0; \\
  2\cdot2 + 4\cdot(-1) = 0; \\
\end{cases}$

Собственные векторы: $\begin{pmatrix}
3\\
-2
\end{pmatrix}$ и $\begin{pmatrix}
2\\
-1
\end{pmatrix}$

**Ответ:** Собственные значения: $\lambda_1 = 3, \lambda_2 = 2$, собственные векторы: $\begin{pmatrix}
3\\
-2
\end{pmatrix}$ и $\begin{pmatrix}
2\\
-1
\end{pmatrix}$

__2.__ Дан оператор поворота на 180 градусов, задаваемый матрицей 

$$A=\begin{pmatrix}
-1 & 0\\ 
0 & -1
\end{pmatrix}.$$

Показать, что __любой__ вектор является для него собственным.

$\begin{pmatrix}
-1 & 0\\ 
0 & -1
\end{pmatrix} \cdot 
\begin{pmatrix}
x\\
y
\end{pmatrix} = 
\lambda \cdot \begin{pmatrix}
x\\
y
\end{pmatrix}$

$\begin{pmatrix}
-x \\ 
-y 
\end{pmatrix} = 
\begin{pmatrix}
\lambda x\\
\lambda y
\end{pmatrix}$

$\begin{cases}
   -x = \lambda x; \\
   -y = \lambda y; \\
 \end{cases}$
 

Собственные значения: $\lambda_{1,2} = -1$

Подставим в уравнение:

$\begin{cases}
   -x = - x; \\
   -y = - y; \\
 \end{cases}$
 
Данной системе удовлетворяют любые координаты => любой вектор является собственным для оператора поворота на 180 градусов.

__3.__ Пусть линейный оператор задан матрицей

$$A=\begin{pmatrix}
1 & 1\\ 
-1 & 3
\end{pmatrix}.$$


Установить, является ли вектор $x=(1,1)$ собственным вектором этого линейного оператора.

$\begin{pmatrix}
1 & 1\\ 
-1 & 3
\end{pmatrix} \cdot 
\begin{pmatrix}
1\\
1
\end{pmatrix} = 
\lambda \cdot \begin{pmatrix}
1\\
1
\end{pmatrix}$

$\begin{cases}
   1 + 1 = \lambda; \\
   -1 + 3 = \lambda;
\end{cases}$
 
$\begin{cases}
   \lambda = 2; \\
   \lambda = 2;
\end{cases}$

Противоречия нет => Вектор $x$ является собственным для линейного оператора, заданного матрицей A, и его собственное значение $\lambda = 2$.

__4.__ Пусть линейный оператор задан матрицей

$$A=\begin{pmatrix}
0 & 3 & 0\\ 
3 & 0 & 0\\
0 & 0 & 3
\end{pmatrix}.$$

Установить, является ли вектор $x=(3, -3, -4)$ собственным вектором этого линейного оператора.

$\begin{pmatrix}
0 & 3 & 0\\ 
3 & 0 & 0\\
0 & 0 & 3
\end{pmatrix} \cdot 
\begin{pmatrix}
3\\
-3\\
-4
\end{pmatrix} = 
\lambda \cdot \begin{pmatrix}
3\\
-3\\
-4
\end{pmatrix}$

$\begin{cases}
   9 = 3\lambda; \\
   -9 = -3\lambda; \\
   -12 = -4\lambda;
\end{cases}$

$\begin{cases}
   \lambda = \frac{1}{3}; \\
   \lambda = \frac{1}{3}; \\
   \lambda = \frac{1}{3};
\end{cases}$

В уравнении не содержится противоречия => Вектор $x$ является собственным для линейного оператора, заданного матрицей A, и его собственное значение = $\lambda = \frac{1}{3}$
